```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

### Import dataset iris

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
num = iris[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
#Convert numerical values to numpy array
X = num.values
X
```

#### Scaling the values

```python
X = scale(X)
X
```

### Apply Principal Component Analysis

```python
pca = PCA(n_components =4)
pca.fit(X)
```

```python
#The amount of variance that each pca explains
var = pca.explained_variance_ratio_
var
```

##### It tells us that 72.96% of the variation in this dataset is explained by the first principal component, 22.85% by second component and so on.

#### Plot the variance

```python
plt.plot(var)
```

Principal component 1 explains the maximum variance and the ability to explain the variance keeps on going down

```python
# Cumulative variance
var1 = np.cumsum(np.round(pca.explained_variance_ratio_ , decimals =4)*100)
```

```python
plt.plot(var1)
```
